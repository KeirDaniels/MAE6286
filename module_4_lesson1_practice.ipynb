```python
import numpy
from matplotlib import pyplot
%matplotlib inline
```

```python
pyplot.rcParams['font.family'] = 'serif'
pyplot.rcParams['font.size'] = 16
```

```python
#set paramerters
L = 1.0
nx = 51 # number of locations on the rod
dx = L / (nx -1) #distance between points

alpha = 1.22e-3
q = 0.0

#define the locations along the rod
x = numpy.linspace(0.0, L, num=nx)

#set the inital temps
T0 = numpy.zeros(nx)
T0[0] = 100.0
```

```python
def ftcs(T0, nt, dt, dx, alpha):
    T = T0.copy()
    sigma = alpha * dt / dx**2
    for n in range(nt):
        
        T[1:-1] = (T[1:-1] +  sigma* (T[2:] - 2.0 *T[1:-1] + T[:-2]))
    

    return T

    
```

```python
#stability constant based on CFL limit is dt * alpha / dx**2 =< 0.5

nt = 100 #time steps 
sigma = 0.5

dt = sigma * dx**2 / alpha #time step

#find temp along the rod
T = ftcs(T0, nt, dt, dx, alpha)
```

```python
#plot everything
pyplot.figure(figsize=(6.0, 4.0))
pyplot.title('Distance [m]')
pyplot.xlabel('Distance [m]')
pyplot.ylabel('Temp [C]')
pyplot.grid()
pyplot.plot(x,T, color='C0', linestyle='-', linewidth=2)
pyplot.xlim(0.0,L)
pyplot.ylim(0.0, 100.0);
```

```python
def ftcd_mixed_bcs(T0, nt, dt, dx, alpha):
    T = T0.copy()
    sigma = dt * alpha / dx**2
    
    for n in range(nt):
        T[1:-1] = T[1:-1] + sigma * (
        T[2:] - 2.0 * T[1:-1] + T[:-2])
        
        T[-1] = T[-2]
    return T
```

```python
nt = 1000

T = ftcd_mixed_bcs(T0, nt, dt, dx, alpha)
```

```python
from scipy import linalg
```

```python
def lhs_operator(N, sigma):
    #set the diag
    D = numpy.diag((2.0 + 1.0/ sigma) * numpy.ones(N))
    #numan condition for th elast element
    D[-1,-1] = 1.0 + 1.0 / sigma
    #the upper diag
    U = numpy.diag(-1.0 * numpy.ones(N-1), k=1)
    #lower diag
    
    L = numpy.diag(-1.0 * numpy.ones(N - 1), k=-1)
    #make the operator
    
    A = D + U + L
    
    return A
    
    
```

```python
def rhs_vector(T, sigma, gdx):
    b = T[1:-1] / sigma
    
    #set dirichlet condition
    
    b[0] += T[0]
    
    #set neumann condition
    
    b[-1] += gdx
    
    return b
```

```python
def btcs_implicit(T0, nt, dt, dx, alpha, q):
    sigma = dt * alpha / dx**2
    
    #creat implicit operator for system
    
    A = lhs_operator(len(T0) - 2, sigma)
    
    #integrate in time
    
    T = T0.copy()
    for n in range(nt):
        #generate rhs
        b = rhs_vector(T, sigma, q*dx)
        
        #solve system 
        
        T[1:-1] = linalg.solve(A,b)
        #apply neuman boundry conditions
        T[-1] = T[-2] + q * dx
        
    return T
```

```python
# Set the time-step size based on CFL limit.
sigma = 0.5
dt = sigma * dx**2 / alpha  # time-step size
nt = 1000  # number of time steps to compute

# Compute the temperature along the rod.
T = btcs_implicit(T0, nt, dt, dx, alpha, q)
```

```python
# Plot the temperature along the rod.
pyplot.figure(figsize=(6.0, 4.0))
pyplot.xlabel('Distance [m]')
pyplot.ylabel('Temperature [C]')
pyplot.grid()
pyplot.plot(x, T, color='C0', linestyle='-', linewidth=2)
pyplot.xlim(0.0, L)
pyplot.ylim(0.0, 100.0);
```
